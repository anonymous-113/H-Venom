```python
import pandas as pd
```

```python
path =r'C:\AI\CollegeWork\H-VENOM_ANNOTATIONS.xlsx'
```

```python
df=pd.read_excel(path)
```

```python
actions_to_check = [
    '1.1. Weapon--Gun', '1.2. Weapon-- Sword', '1.3. Weapon-- Arrow', '1.4. Weapon-- Knife', '2. Fire', '3. Abusive Language',
    '4. Blood/ Murder', '5. Physical Fighting', '6. Racism', '7. Hate towards Women',
    '8. Hate towards Child', '9. Other'
]

# Initialize a dictionary to store counts for each action
action_counts = {action: 0 for action in actions_to_check}

# Process each row in the DataFrame
for action_str in df['Action2']:
    # Split the action string by commas and strip extra spaces
    actions = [action.strip() for action in action_str.split(',') if action.strip()]
    # Increment counts for each action in the list
    for action in actions:
        if action in action_counts:
            action_counts[action] += 1

# Display the result
print("Number of occurrences for each Action2 label:")
for action, count in action_counts.items():
    print(f"{action}: {count}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# Data for the number of occurrences for each Action2 label
action_counts = {
    '1.1. Weapon--Gun': ,
    '1.2. Weapon-- Sword': ,
    '1.3. Weapon-- Arrow': ,
    '1.4. Weapon-- Knife': ,
    '2. Fire': ,
    '3. Abusive Language': ,
    '4. Blood/ Murder': ,
    '5. Physical Fighting': ,
    '6. Racism': ,
    '7. Hate towards Women': ,
    '8. Hate towards Child': ,
    '9. Other': 
}

# Define a custom color map from skyblue to salmon
def create_gradient_cmap(start_color, end_color, n_colors):
    """Creates a gradient color map from start_color to end_color with n_colors."""
    start_rgb = np.array(mcolors.to_rgb(start_color))
    end_rgb = np.array(mcolors.to_rgb(end_color))
    gradient = np.linspace(start_rgb, end_rgb, n_colors)
    return mcolors.ListedColormap(gradient)

# Create a vertical bar chart
plt.figure(figsize=(16, 10))

# Create gradient colors from skyblue to salmon
gradient_cmap = create_gradient_cmap('skyblue', 'salmon', len(action_counts))

bars = plt.bar(list(action_counts.keys()), action_counts.values(), color=gradient_cmap(np.linspace(0, 1, len(action_counts))), edgecolor='black')

# Add labels and title
plt.ylabel('Number of Occurrences', fontsize=14, fontweight='bold', color='black')
plt.title('Number of Occurrences for Each Action2 Label', fontsize=16, fontweight='bold', color='black')

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 10, f'{height}', 
             ha='center', va='bottom', fontsize=12, color='black', weight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=12, fontweight='bold', color='black')

# Add a horizontal line at the top of the bars for visual effect
plt.axhline(y=max(action_counts.values()) + 30, color='grey', linestyle='--', linewidth=0.7)

# Set x-axis limits to provide space for labels
plt.xlim(-0.5, len(action_counts) - 0.5)

# Show the plot
plt.tight_layout()
plt.show()
```
