```python
import os
import json
import assemblyai as aai

# Define paths
audio_dir = ""  # Directory containing mono audio files
output_file = "."  # JSON file to save transcription data

# Initialize Assembly AI client
aai.settings.api_key = ""
transcriber = aai.Transcriber()
```

```python
def transcribe_audio(audio_path):
    try:
        # Transcribe audio
        transcript = transcriber.transcribe(audio_path)
        
        # Wait for transcription to complete
        while transcript.status != aai.TranscriptStatus.completed:
            if transcript.status == aai.TranscriptStatus.error:
                print(f"Error: {transcript.error}")
                return None
            transcript = transcriber.get_transcript(transcript.id)
        
        # Return the transcription text
        return transcript.text
    except Exception as e:
        print(f"Error processing {audio_path}: {e}")
        return None

def append_to_json(data):
    try:
        # Read existing data
        if os.path.exists(output_file):
            with open(output_file, 'r') as f:
                existing_data = json.load(f)
        else:
            existing_data = []

        # Append new data
        existing_data.append(data)

        # Write back to file
        with open(output_file, 'w') as f:
            json.dump(existing_data, f, ensure_ascii=False, indent=4)
    except Exception as e:
        print(f"Error appending to JSON file: {e}")

def append_to_json(data, output_file):
    try:
        # Read existing data
        if os.path.exists(output_file):
            with open(output_file, 'r', encoding='utf-8') as f:
                try:
                    existing_data = json.load(f)
                except json.JSONDecodeError:
                    existing_data = []
        else:
            existing_data = []

        # Append new data
        existing_data.append(data)

        # Write back to file
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(existing_data, f, ensure_ascii=False, indent=4)
    except Exception as e:
        print(f"Error appending to JSON file: {e}")
# Process each audio file
for filename in os.listdir(audio_dir):
    if filename.endswith('.wav'):
        audio_path = os.path.join(audio_dir, filename)
        video_name = os.path.splitext(filename)[0]
        transcription = transcribe_audio(audio_path)
        if transcription:
            result = {
                "video_name": video_name,
                "transcription": f"{transcription}"
            }
        append_to_json(result,output_file)
        print(f"Transcribed and appended {filename}")

print("Transcription data appended to", output_file)
```
